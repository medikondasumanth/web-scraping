```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
def Product_name(soup):

    try:
        
        title = soup.find("span", attrs={"id":'productTitle'})
        
        title_value = title.text

        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string
```

```python
def User_rating(soup):

    try:
        rating = soup.find("i", attrs={'class':'a-icon a-icon-star a-star-4-5'}).string.strip()
    
    except AttributeError:
        try:
            rating = soup.find("span", attrs={'class':'a-icon-alt'}).string.strip()
        except:
            rating = ""

    return rating


```

```python
def User_reviews(soup):
    try:
        review_count = soup.find("span", attrs={'id':'acrCustomerReviewText'}).string.strip()

    except AttributeError:
        review_count = ""

    return review_count
```

```python
def Product_description(soup):
    try:
        description = soup.find("div", attrs={"id": 'feature-bullets'})
        if description:

            descriptionpoints = [point.get_text(strip=True) for point in description.find_all("span", attrs={"class": 'a-list-item'})]
        else:
            descriptionpoints = []
    except AttributeError:
        descriptionpoints = []

    return descriptionpoints
```

```python
def Product_imageurl(soup):
    try:
        image = soup.find("div", attrs={"id": "imgTagWrapperId"})
        if image:
            image_url = image.find("img")["src"]
        else:
            image_url = ""
    except (AttributeError, KeyError):
        image_url = ""

    return image_url
```

```python
def Product_description_1(soup):
    try:
        description = soup.find("div", attrs={"id":'productFactsDesktop_feature_div'})
        if description:
            
            descriptionpoints = [point.get_text(strip=True) for point in description.find_all("span", class_='a-list-item a-size-base a-color-base')]
        else:
            descriptionpoints = []
    except AttributeError:
        descriptionpoints = []

    return descriptionpoints
```

```python
def Product_price(soup):
    try:
        price = soup.find("span", attrs={'class':'a-price-whole'})
        if price:
            price_value = price.text.strip()
            
            price_value = "₹" + price_value
        else:
            price_value = ""
    except AttributeError:
        price_value = ""
    return price_value

```

```python
if __name__ == '__main__':

    headers = ({'User-Agent':'', 'Accept-Language': 'en-US, en;q=0.5'})

    url = "https://www.amazon.in/s?k=smartwatchs&crid=24LZSRMX64HCD&sprefix=smartwatchs%2Caps%2C224&ref=nb_sb_noss_2"

    response = requests.get(url, headers=headers)
    print(response)

    soup = BeautifulSoup(response.content, "html.parser")

    links = soup.find_all("a", attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
                          
    links_list = []

    # Loop for extracting links from Tag Object
    for link in links:
            links_list.append(link.get('href'))

    dic = { 
            "Name":[], 
            "Price":[], 
            "Description":[],
            "Reviews":[], "Rating":[],
            "Image URL":[]
        }

    
    # Loop for extracting product details from each link 
    for link in links_list:
        new_response= requests.get("https://www.amazon.in" + link, headers=headers)

        new_soup = BeautifulSoup(new_response.content, "html.parser")

        # Function calls to display all necessary product information
        dic['Name'].append(Product_name(new_soup))
        dic['Price'].append(Product_price(new_soup))
        dic['Description'].append(Product_description(new_soup))
        dic['Reviews'].append(User_reviews(new_soup))
        dic['Rating'].append(User_rating(new_soup))
        dic['Image URL'].append(Product_imageurl(new_soup))

    
    amazon_df = pd.DataFrame.from_dict(dic)
    amazon_df['Name'].replace('', np.nan, inplace=True)
    amazon_df = amazon_df.dropna(subset=['Name'])
    amazon_df = amazon_df.dropna(subset=['Price'])
    amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
